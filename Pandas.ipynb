```python
# # DATA VISUALIZATION USING pandas
# Import the python libraries

import pandas as pd 
import matplotlib.pyplot as plt
import os
```

```python
os.getcwd()
```

```python
 # Load the Iris dataset

iris = pd.read_csv("Iris.csv")
print(iris.head())
```

```python
# Load the Wine reviews dataset

wine_reviews = pd.read_csv('winemag-data-130k-v2.csv', index_col=0)
wine_reviews.head()
```

```python
# Scatter plot

iris.plot.scatter(x='SepalLengthCm', y='SepalWidthCm', title='Scatter plot', s=35, figsize=(8,6))
plt.show()
```

```python
# Line chart

iris.drop(['Id','Species'], axis=1).plot.line(title='Line plot', figsize=(8,6))
plt.show()
```

```python
# Area chart

iris.drop(['Id', 'Species'], axis=1).plot.area(title='Area chart', colormap='Spectral', figsize=(8, 8))
plt.show()
```

```python
# Pie chart

wine_reviews.groupby("country").price.mean().sort_values(ascending=False)[:5].plot.pie(title='Pie chart',colormap='cividis',autopct='%1.2f%%',figsize=(7,7))
plt.show()
```

```python
# Bar chart

wine_reviews.groupby("country").points.sum().sort_values(ascending=True)[:10].plot.bar(title='Bar chart', color='Green', figsize=(12,8))
plt.show()
```

```python
# Histogram

wine_reviews['points'].plot.hist(bins=30, color='Green',title='Histogram plot')
plt.show()
```

```python
# Box plot

iris.drop("Id", axis=1).boxplot(by="Species", figsize=(10, 10), color='Red')
plt.show()
```

```python
# Scatter matrix

from pandas.plotting import scatter_matrix

fig, ax = plt.subplots(figsize=(12,12))
scatter_matrix(iris.drop("Id", axis=1), ax=ax)
plt.show()
```

```python
# Example 1
# Scatter plot with multiple colors

colors = {'Iris-setosa':'red', 'Iris-versicolor':'green', 'Iris-virginica':'blue'}
fig, ax = plt.subplots(figsize=(8,8))
grouped = iris.drop(['Id'], axis=1).groupby('Species')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='SepalLengthCm', y='SepalWidthCm', label=key, color=colors[key], title = 'Scatter plot grouped by Species')
plt.show()
```

```python
# Example 2
# Scatter plot with markers

iris.plot.scatter(x="SepalLengthCm", y="SepalWidthCm", color="red", marker="*", s=150, figsize=(8,6), title = 'Scatter plot with triangles')
plt.show()
```

```python
# Example 3
# Line plot with subplots 

iris.drop(['Id','Species'], axis=1).plot(subplots=True, title = 'Scatter plot with subplots', figsize=(8, 8));
```

```python
# Example 4
# Histogram for multiple columns

wine_reviews.hist(layout=(1,2), figsize=(16,8))
plt.show()
```

```python
# Example 5
# Bar chart with bins

wine_reviews['points'].value_counts().sort_index().plot.bar(color='orange', title='Frequency distribution of wine points')
plt.show()
```

```python
# Example 6
# Horizontal bar chart

wine_reviews['points'].value_counts().sort_index().plot.barh(title = 'Horizontal frequency distribution of wine points')
plt.show()
```

```python
# Example 7
# Bar chart with country wise average wine prices

wine_reviews.groupby("country").price.mean().plot(kind="bar", title='Mean wine prices of different countries', figsize=(16, 8), colormap='Spectral')
plt.show()
```

```python
# Example 8
# Bar chart with groups

iris.drop(['Id','Species'], axis=1).sort_values(by='SepalWidthCm', ascending=True)[:5].plot(kind='bar', colormap='plasma',title='Grouped bar chart', figsize=(8, 8));
```

```python
# Example 9
# Stacked bar chart

iris.drop(['Id','Species'], axis=1).sort_values(by='SepalLengthCm', ascending=False)[:5].plot.bar(stacked=True,title='Stacked bar chart',colormap='inferno',figsize=(8, 8));
```

```python
# Example 10
# Horizontal stacked bar chart

iris.drop(['Id','Species'], axis=1).sort_values(by='SepalLengthCm', ascending=False)[:5].plot.barh(stacked=True,title='Horizontal stacked bar chart',colormap='magma',figsize=(8, 8));
```
